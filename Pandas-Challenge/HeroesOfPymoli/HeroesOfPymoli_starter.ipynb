## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns

* Create a summary data frame to hold the results

* Display a preview of the summary data frame



```python
Popular_Items = purchase_data[['Item ID', 'Item Name', 'Price']]
Popular_Items #test Succesfully new data frame printed with the information asked
```

```python
#Using the group by and the rename function in the Item ID and Item Name variable,we will get the purchase cout, Avg Purchase Price, Avg Total Purchases Total and the purchase total per person 
Popular_Items_Analysis = Popular_Items.groupby(["Item ID","Item Name"]).count()["Price"].rename("Purchase Count")
#Popular_Items_Analysis #Test Run Succesfully (0 Splinter 4, 1 Crucifer 3, 2 Verdict 6, 3 Phantomlight 6,4 Bloodlord's Fetish 5,...)
Ave_PopularItems_Price = Popular_Items.groupby(["Item ID","Item Name"]).mean()["Price"].rename("Average Purchase Price")
#Ave_PopularItems_Price #Test Run Succesfully (0 Splinter 1.28,1 Crucifer 3.26, 2 Verdict 2.48, 3 Phantomlight 2.49, 4 Bloodlord's Fetish 1.70,...)
Age_total_PopularItems_Val = Popular_Items.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Purchase Value")
#Age_total_PopularItems_Val #Test Run Succesfully (0 Splinter 5.12,1 Crucifer 9.78, 2 Verdict 14.88, 3 Phantomlight 14.94, 4 Bloodlord's Fetish 8.50,...)
```

```python
#Create a Data frame that includes this values
Pop_Items = pd.DataFrame({"Purchase Count": Popular_Items_Analysis,
                              "Average Purchase Price": Ave_PopularItems_Price,
                              "Total Purchase Value": Age_total_PopularItems_Val
                            })
Pop_Items.head(6) #Test Succesfuly Data frame printed
```

```python
#Sort the order of the data frame so it goes from Purchase Count from top to bottom
PopItems_df = Pop_Items.sort_values(by='Purchase Count', ascending=False)
PopItems_df.head(6) #Test Sucesfully new data frame printed
```

```python
#Giving format to the Average Purchase Price and the Total Value Purchase
PopItems_df["Average Purchase Price"] = PopItems_df["Average Purchase Price"].map("${:.2f}".format)
PopItems_df["Total Purchase Value"] = PopItems_df["Total Purchase Value"].map("${:.2f}".format)
#Test Sucesfully the data was sucesfully format
```
